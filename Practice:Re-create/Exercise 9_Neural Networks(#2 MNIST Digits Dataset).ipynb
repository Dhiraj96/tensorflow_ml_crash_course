### Some background history

The National Institute of Standards and Technology (NIST) is a US federal agency specifically tasked with developing measurement standards. The agency has put together several datasets of images showing handwritten digits. These were standardized and put together by Yann LeCun and Corinna Cortes while at AT&T Labs into a classification dataset. The prediction task is to recognize what number is shown in the image.

The dataset, known as MNIST or MNIST-10, is a very commonly used training set when learning how to do image recognition. In academic research, it is certainly overused, but it does make a great example for teaching purposes. The images are relatively small but still represent an interesting classification task. The primary reason for this is that digit classification does not require color or high-resolution images, making the data size relatively small.

### Sanity Check

Lets parse out the labels and features seperately for ease once we pass these values to our model. 

```python
"""Extract and scale/transform the features as needed

    Args:
        dataset: A pandas 'Dataframe', containing labels and feature values 
        /pixel intensity of our images
    Returns:
        'tuple' (labels, features):
            labels: A pandas 'Series'
            features: A pandas 'Dataframe'
"""
def parse_labels_and_features(dataset):
    
    labels = dataset[0]
    
    #The loc allows us to pull out columns based on location without header rows
        # we pull out every row and columns 1 through 784 inclusive
    features = dataset.loc[:,1:784]
    #scale features to be between [0,1] by dividing by max value
    features = features / 255
    
    return labels, features
```

### First things first...

Lets try and create a baseline model to compare against. We will use the LinearClassifier object that provides a set of k one vs. all classifiers for k classes (10 in this case). This essentially means that there are 10 seperate binary classifiers for each example which the model trains seperately and applies to our dataset to reach a final prediction. 

In addition to reporting accuracy and log-loss plotting, we'll also look at something called a confusion matrix. This will show us what classes were misclassified as other classes (what digits got confused for each other). 

```python
"""Constructs and defines the feature columns to use in TF

    Returns: 
        A 'set' of feature columns
"""
def construct_feature_columns():
    
    return set([tf.feature_column.numeric_column('pixels', shape = 784)])
```

Since we will need seperate input functions for training and prediction, we will nest them in `create_training_input_fn()` and `create_predict_input_fn()` respectively. By invoking these functions, we return the corresponding `_input_fn`s to pass where needed (during our .train() and .predict() calls). 

The following two functions are named closure functions. These functions have the following features:

    1. A nested function
    2. The nested function refers to a value defined in the enclosing  function
    3. the enclosing funtion must return the nested function
    
Why do we use these functions though? We avoid the use of global values and provide some form of data hiding (a more OOP approach). Instead of using an actual class for function polymorphism, using closures provide a more elegant solution

```python
"""Custom input function to send MNIST data to TF estimator for training

    Args:
        features: Training features
        labels: Training labels
        batch_size: Batch size to use during training
    Returns:
        Function that returns batches of training features and labels 
        to use during training

"""
def create_training_input_func(features, labels, batch_size, num_epochs=None, shuffle = False):
        
    def _input_func(num_epochs = None, shuffle=True):
        
        # To ensure model gets a good sampling of data (even with small steps)
        # we shuffle all the data before creating a dataset.
        # Remember that input pipelines reset with each call to .train()
        idx = np.random.permutation(features.index)
        raw_features = {"pixels":features.reindex(idx)}
        raw_targets = np.array(labels[idx])
        
        ds = Dataset.from_tensor_slices((raw_features, raw_targets))
        ds = ds.batch(batch_size).repeat(num_epochs)    
        
        if shuffle:
            ds = ds.shuffle(10000)
            
        #return next batch of data
        feature_batch, label_batch = ds.make_one_shot_iterator().get_next()
        return feature_batch, label_batch
    
    return _input_func
```

```python
"""Custom input function to send MNIST data to TF estimator for predictions

    Args:
        features: Features to base predictions on
        labels: labels of prediction examples
    Returns:
        A function returning feature and labels for predictions

"""
def create_predict_input_func(features, labels, batch_size):
    
    def _input_func():
        
        raw_features = {"pixels": features.values}
        raw_targets = np.array(labels)
        
        ds = Dataset.from_tensor_slices((raw_features, raw_targets))
        ds = ds.batch(batch_size)
        
        #return next batch of data
        feature_batch, label_batch = ds.make_one_shot_iterator().get_next()
        return feature_batch, label_batch
    
    return _input_func
    
    
```

```python
"""Trains 10-way linear classification model for the MNIST digits dataset. 

    Args:
        learning_rate: A 'float', learning rate
        steps: A non-zero 'int', total training steps taken (each step is forward
            and backward pass using single batch)
        batch_size: A non-zero 'int', the batch size
        training_examples: A `DataFrame` containing input features for training.
        training_targets: A `DataFrame` containing targets for training.
        validation_examples: A `DataFrame` containing input features for validation.
        validation_targets: A `DataFrame` containing targets for validation.
            
    Returns:
        A trained 'LinearClassifier' object trained on training data
        Plots RMSE at each training period step for training and validation data
"""
def train_linear_classification_model(learning_rate, steps, batch_size,
                                      training_examples, training_targets,
                                      validation_examples, validation_targets):
    
    periods = 10
    steps_per_period = steps/periods
    
    #input functions
    training_input_func = create_training_input_func(
        training_examples, training_targets, batch_size)
    
    predict_training_input_func = create_predict_input_func(
        training_examples, training_targets, batch_size)
    
    predict_validation_input_func = create_predict_input_func(
        validation_examples, validation_targets, batch_size)
    
    #Create linear classifier object
    my_optimizer = tf.train.AdagradOptimizer(learning_rate=learning_rate)
    my_optimizer = tf.contrib.estimator.clip_gradients_by_norm(my_optimizer, 5.0)
    
    classifier = tf.estimator.LinearClassifier(
        feature_columns = construct_feature_columns(),
        n_classes = 10,
        optimizer = my_optimizer,
        config = tf.estimator.RunConfig(keep_checkpoint_max=1)
    )
    
    #Train model inside loop to periodically asses loss metrics
    print("Training Model...")
    print("Logloss error (validation data): ")
    training_errors = []
    validation_errors = []
    
    for period in range(0, periods):
        
        classifier.train(
            input_fn = training_input_func,
            steps = steps_per_period
        )
        
        #compute probabilities 
        training_predictions = list(classifier.predict(input_fn = predict_training_input_func))
        training_probabilities = np.array([item['probabilities'] for item in training_predictions])
        training_pred_class_id = np.array([item['class_ids'][0] for item in training_predictions])
        training_pred_one_hot = tf.keras.utils.to_categorical(training_pred_class_id, 10)
        
        validation_predictions = list(classifier.predict(input_fn = predict_validation_input_func))
        validation_probabilities = np.array([item['probabilities'] for item in validation_predictions])
        validation_pred_class_id = np.array([item['class_ids'][0] for item in validation_predictions])
        validation_pred_one_hot = tf.keras.utils.to_categorical(validation_pred_class_id, 10)
        
        #compute training and validation errors and append to lists
        training_log_loss = metrics.log_loss(training_targets, training_pred_one_hot)
        training_errors.append(training_log_loss)
        
        validation_log_loss = metrics.log_loss(validation_targets, validation_pred_one_hot)
        validation_errors.append(validation_log_loss)
        
        #print current loss
        print("  period %02d : %0.2f" % (period, validation_log_loss))
        
    print("Model training complete!")
    
    #remove event files to save disk space
    trash = map(os.remove, glob.glob(os.path.join(classifier.model_dir, 'events.out.tfevents*')))
    
    #calculate final predictions (not probabilities as we did above)
    final_predictions = classifier.predict(input_fn = predict_validation_input_func)
    final_predictions = np.array([item['class_ids'][0] for item in final_predictions])
    
    accuracy = metrics.accuracy_score(validation_targets, final_predictions)
    print("Final classifier accuracy: %0.2f" % (accuracy))
    
    # Output a graph of loss metrics over periods.
    plt.ylabel("LogLoss")
    plt.xlabel("Periods")
    plt.title("LogLoss vs. Periods")
    plt.plot(training_errors, label="training")
    plt.plot(validation_errors, label="validation")
    plt.legend()
    plt.show()
    
    #Calculate & normalize Confusion matrix by row
    conf_mat = metrics.confusion_matrix(validation_targets, final_predictions)
    conf_mat_norm = conf_mat.astype("float") / conf_mat.sum(axis=1)[:, np.newaxis]
    
    ax = sns.heatmap(conf_mat_norm, cmap="bone_r")
    ax.set_aspect(1)
    plt.title("Confusion matrix")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()
    
    return classifier
    
```

I spent about 5-10 minutes tweaking hyperparameters for batch size, learning rate and steps to get the following model. 

```python
classifier = train_linear_classification_model(
    learning_rate = 0.01,
    steps = 1000,
    batch_size = 50,
    training_examples = training_examples,
    training_targets = training_targets,
    validation_examples = validation_examples,
    validation_targets = validation_targets)
```

Before moving on, I do want to explain a few other regularization/tuning methods we can use for neural networks that I think are beneficial to know. I know, this post is exposition heavy but bear with me! Neural networks are a powerful tool but are very easy to misuse. 

When I started out, I (naively) spent hours brute forcing or inefficiently grid searching for the optimal set of hyperparameters only to realize I grossly overfit on the training data. 

#### Tuning Neural Networks

1. Dropout Rate/Layer: This is one way to regularize a neural network. This layer (only during training) randomly converts a proportion of its inputs to zero. This forces the next layer to smooth out weights over all the inputs in the prior layer. Essentially, it forces the model to spread out over correlated values. Using a rate of 0.5 is usually a good start. 

2. Its always a good place to start with 2 hidden layers (64 nodes each), a learning rate of 0.01 and make all hidden layers the same number of nodes. You can then change this accordingly based on idiosyncracies in your respective dataset. 

3. We can then change the following parameters as you test which ones work best. Make sure to change one thing at a time and run enough steps to get the model to roughly converge:

      a.) experiment with between 2 and 5 hidden layers
    
      b.) try doubling or halving hidden nodes in powers of 2. The backpropogation algorithm is non-stochastic and therefore noisy so any change in performance you see by slighly changing the number of hidden nodes is likely to be just noise
    
      c.) If the model is very noisy (depends on dataset), then we can decrease the dropout rate (halving). If overfitting, then increase the dropout rate (by 0.1-0.15 each time)
    
Keep in mind, this barely scratches the surface in neural network tuning but I think is a good place to start. 


```python
"""This function trains a neural network classification model for the MNIST digits dataset

    Args:
        learning_rate: A 'float', learning rate
        steps: A non-zero 'int', total training steps taken (each step is forward
            and backward pass using single batch)
        batch_size: A non-zero 'int', the batch size
        hidden_units: A 'list' of int values, specifying neurons in each layer
        training_examples: A `DataFrame` containing input features for training.
        training_targets: A `DataFrame` containing targets for training.
        validation_examples: A `DataFrame` containing input features for validation.
        validation_targets: A `DataFrame` containing targets for validation.
            
    Returns:
        A 'DNNClassifier' object trained on training data
        Plots LogLoss at each training period step for training and validation data as 
        as well as a confusion matrix 

"""
def train_nn_classification_model(learning_rate, steps, batch_size, hidden_units, 
                                  training_examples, training_targets,
                                  validation_examples, validation_targets):
    
    #If the number of steps is too small, your model may never see most of the data
    #We can rectify this by manually controlling length of training using num_epochs
    #in our training/predict input functions or shuffling all the data. 
        #(This is why we have those unused arguments in the 2 aforementioned functions)
    periods = 10
    steps_per_period = steps/periods
    
    #input functions
    training_input_func = create_training_input_func(
        training_examples, training_targets, batch_size)
    
    predict_training_input_func = create_predict_input_func(
        training_examples, training_targets, batch_size)
    
    predict_validation_input_func = create_predict_input_func(
        validation_examples, validation_targets, batch_size)
    
    #create feature columns
    feature_columns = [tf.feature_column.numeric_column('pixels', shape=784)]
    
    #Create DNNClassifier object
    my_optimizer = tf.train.AdagradOptimizer(learning_rate = learning_rate)
    my_optimizer = tf.contrib.estimator.clip_gradients_by_norm(my_optimizer, 5.0)
    
    classifier = tf.estimator.DNNClassifier(
        feature_columns = feature_columns,
        n_classes = 10,
        hidden_units = hidden_units,
        optimizer = my_optimizer,
        dropout = 0.5,
        config = tf.contrib.learn.RunConfig(keep_checkpoint_max=1)
    )
    
    
    print("Training model...")
    print("LogLoss error (on validation data)")
    training_errors = []
    validation_errors = []
    
    for period in range(0, periods):
        
        classifier.train(
            input_fn = training_input_func,
            steps = steps_per_period
        )
        
        #compute probabilities 
        training_predictions = list(classifier.predict(input_fn = predict_training_input_func))
        training_probabilities = np.array([item['probabilities'] for item in training_predictions])
        training_pred_class_id = np.array([item['class_ids'][0] for item in training_predictions])
        training_pred_one_hot = tf.keras.utils.to_categorical(training_pred_class_id, 10)
        
        validation_predictions = list(classifier.predict(input_fn = predict_validation_input_func))
        validation_probabilities = np.array([item['probabilities'] for item in validation_predictions])
        validation_pred_class_id = np.array([item['class_ids'][0] for item in validation_predictions])
        validation_pred_one_hot = tf.keras.utils.to_categorical(validation_pred_class_id, 10)
        
        #compute training and validation errors and append to lists
        training_log_loss = metrics.log_loss(training_targets, training_pred_one_hot)
        training_errors.append(training_log_loss)
        
        validation_log_loss = metrics.log_loss(validation_targets, validation_pred_one_hot)
        validation_errors.append(validation_log_loss) 
        
        #print current loss
        print("  period %02d : %0.2f" % (period, validation_log_loss))
        
    print("Model training complete!")
    
    #remove event files to save disk space
    trash = map(os.remove, glob.glob(os.path.join(classifier.model_dir, 'events.out.tfevents*')))
    
    #calculate final predictions (not probabilities as we did above)
    final_predictions = classifier.predict(input_fn = predict_validation_input_func)
    final_predictions = np.array([item['class_ids'][0] for item in final_predictions])
    
    accuracy = metrics.accuracy_score(validation_targets, final_predictions)
    print("Final classifier accuracy: %0.2f" % (accuracy))
    
    # Output a graph of loss metrics over periods.
    plt.ylabel("LogLoss")
    plt.xlabel("Periods")
    plt.title("LogLoss vs. Periods")
    plt.plot(training_errors, label="training")
    plt.plot(validation_errors, label="validation")
    plt.legend()
    plt.show()
    
    #Calculate & normalize Confusion matrix by row
    conf_mat = metrics.confusion_matrix(validation_targets, final_predictions)
    conf_mat_norm = conf_mat.astype("float") / conf_mat.sum(axis=1)[:, np.newaxis]
    
    ax = sns.heatmap(conf_mat_norm, cmap="bone_r")
    ax.set_aspect(1)
    plt.title("Confusion matrix")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
    plt.show()
    
    return classifier
    
    
    
```

```python
nn_classifier = train_nn_classification_model(
    learning_rate = 0.015,
    steps = 2000,
    batch_size = 50,
    hidden_units = [128,128],
    training_examples = training_examples,
    training_targets = training_targets,
    validation_examples = validation_examples,
    validation_targets = validation_targets)
```

### Visualizing weights (Neural Network Internals)

Now, this part may be a bit confusing (unless you watched the video I linked at the start of this exercise and in the last two). 



```python
print(nn_classifier.get_variable_names())

weights0 = nn_classifier.get_variable_value("dnn/hiddenlayer_0/kernel")

print("weights0 shape:", weights0.shape)

num_nodes = weights0.shape[1]
num_rows = int(math.ceil(num_nodes / 10.0))
fig, axes = plt.subplots(num_rows, 10, figsize=(20,2*num_rows))
for coef, ax in zip(weights0.T, axes.ravel()):
    
    #reshape weights in coef from 1x784 to 28x28
    ax.matshow(coef.reshape(28,28), cmap=plt.cm.pink)
    ax.set_xticks(())
    ax.set_yticks(())
    
plt.show()
```

As we can see, the first hidden layer is modelling some low level fuzzy features of the digits. It's not surprising to also see some neurons that are essentially just noise (especially with our dropout rate). This is usually an indication that they have not converged or are being ignored by higher layers.
